```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Sample dataset
# Each row represents a transaction with features: amount, time_since_last_transaction, and distance_from_home
data = {
    "Transaction_Amount": [50, 20, 30, 1000, 25, 15, 2000, 40, 10, 35],
    "Time_Since_Last_Transaction": [5, 3, 8, 1, 7, 2, 0.5, 6, 2, 4],
    "Distance_From_Home": [2, 1.5, 2.1, 15, 1.8, 2.5, 30, 1.6, 1.4, 2.2],
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Step 1: Normalize the data
scaler = StandardScaler()
scaled_data = scaler.fit_transform(df)

# Step 2: Apply K-Means Clustering
kmeans = KMeans(n_clusters=2, random_state=42)  # Assume we are clustering into 2 groups
df['Cluster'] = kmeans.fit_predict(scaled_data)

# Step 3: Calculate the distance of each point to its cluster center
df['Distance_To_Center'] = np.min(kmeans.transform(scaled_data), axis=1)

# Step 4: Set a threshold for fraud detection (e.g., top 10% of distances)
threshold = df['Distance_To_Center'].quantile(0.9)
df['Is_Fraud'] = df['Distance_To_Center'] > threshold

# Display results
print("Cluster Centers:\n", kmeans.cluster_centers_)
print("\nData with Clusters and Fraud Flag:\n", df)

# Step 5: Visualize the clusters and anomalies
plt.figure(figsize=(8, 6))
plt.scatter(df['Transaction_Amount'], df['Distance_From_Home'], c=df['Cluster'], cmap='viridis', s=100)
fraud_points = df[df['Is_Fraud']]
plt.scatter(fraud_points['Transaction_Amount'], fraud_points['Distance_From_Home'], c='red', label='Fraud', s=200, marker='x')
plt.xlabel('Transaction Amount')
plt.ylabel('Distance From Home')
plt.title('K-Means Clustering for Fraud Detection')
plt.legend()
plt.show()
```
